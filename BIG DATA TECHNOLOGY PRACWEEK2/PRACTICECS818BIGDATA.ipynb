```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv("train.csv")
```

```python
titanic.plot(kind='box',subplots=True,layout=(3,3))
plt.show()
```

```python
titanic['Survived'].value_counts().plot.bar()
plt.title('Titanic survivor vs fartiles')
plt.show()
```

```python
titanic['Pclass'].value_counts().sort_values(ascending=False).plot.barh()
plt.title('Passenger classes')
plt.show()
```

```python
titanic['SibSp'].plot.density()
plt.show()
```

```python
titanic.groupby(['Pclass','Sex']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

```python
corr = titanic.corr(numeric_only = True)
sns.heatmap(corr)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(titanic["Age"], titanic["Pclass"], color='g', marker='^')
ax.scatter(titanic["Age"], titanic["SibSp"], color='y', marker = '*')
plt.show()
```
